```python
import tkinter as tk
from tkinter import messagebox

departments = ["Computer Science", "Mass Communication"]
admitted = []
not_admitted = []


def show_results():
    if admitted:
        admitted_list = "\n".join([f"{name} - {dept}" for name, dept in admitted])
    else:
        admitted_list = "None"

    if not_admitted:
        not_admitted_list = "\n".join([f"{name} - {dept}" for name, dept in not_admitted])
    else:
        not_admitted_list = "None"

    message = f"Admitted Students:\n{admitted_list}\n\nNot Admitted Students:\n{not_admitted_list}"
    messagebox.showinfo("Admission Results", message)



def requirements(department):
    score = int(score_entry.get().strip())
    credit = int(credit_entry.get().strip())
    interview = interview_entry.get().strip().lower()
    name = name_entry.get().strip().title()
    
    if department == departments[0]:
        if score >= 250 and credit >=  5 and interview == 'y':
            admitted.append((name, department))
            message = f"Congrats {name} you have been admitted into {department.title()}"
            messagebox.showinfo("Congrats", message)
        else :
            not_admitted.append((name, department))
            message = f"Sorry {name} you have not been admitted into {department.title()}"
            messagebox.showinfo("Sorry", message)
            
    if department == departments[1]:
        if score >= 230 and credit >=  5 and interview == 'y':
            admitted.append((name, department))
            message = f"Congrats {name} you have been admitted into {department.title()}"
            messagebox.showinfo("Congrats", message)
        else :
            not_admitted.append((name, department))
            message = f"Sorry {name} you have not been admitted into {department.title()}"
            messagebox.showinfo("Sorry", message)
        

def student_department():
    department = department_entry.get().strip().title()

    if department in departments:
        requirements(department)
    else:
        message = f"Sorry we don't have this department {department.title()}"
        messagebox.showinfo("Sorry", message)




root = tk.Tk()
root.title("Admission checker")
root.geometry("700x500")

tk.Label(root, text="Enter your name:").pack(pady=(20, 10))
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter your Department:").pack(pady=(20, 10))
department_entry = tk.Entry(root)
department_entry.pack()

tk.Label(root, text="Enter your JAMB score:").pack(pady=(20, 10))
score_entry = tk.Entry(root)
score_entry.pack()

tk.Label(root, text="How many Credits do you have:").pack(pady=(20, 10))
credit_entry = tk.Entry(root)
credit_entry.pack()

tk.Label(root, text="Did you pass the interview, type Y or N: ").pack(pady=(20, 10))
interview_entry = tk.Entry(root)
interview_entry.pack()

tk.Button(root, text="Submit", command=student_department).pack(pady=20)
tk.Button(root, text="Show Results", command=show_results).pack(pady=10)


root.mainloop()
```
